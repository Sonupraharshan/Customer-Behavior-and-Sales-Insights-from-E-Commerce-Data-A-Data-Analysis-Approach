```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = ['Warehouse_block','Mode_of_Shipment','Product_importance', 'Gender']
for i in cols:
    le.fit(df[i])
    df[i] = le.transform(df[i])
    print(i, df[i].unique())
```

```python
sns.violinplot(x = 'Customer_care_calls', y = 'Cost_of_the_Product', data = df)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('Reached.on.Time_Y.N', axis=1), df['Reached.on.Time_Y.N'], test_size=0.2, random_state=10)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
from sklearn.cluster import KMeans
n_clusters=2
kmeans = KMeans(n_clusters=n_clusters, random_state=10)
X_train['Cluster'] = kmeans.fit_predict(X_train_pca)
X_test['Cluster'] = kmeans.predict(X_test_pca)  
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, mean_absolute_error, r2_score, mean_squared_error
knn = KNeighborsClassifier()
knn.fit(X_train_pca, y_train)
print("accuracy:",knn.score(X_train_pca, y_train)*100)
```

```python
knn_pred = knn.predict(X_test_pca)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_pca)
X_test_scaled = scaler.transform(X_test_pca)

lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)
print("Training accuracy:", lr.score(X_train_scaled, y_train) * 100)
```

```python
lr_pred = lr.predict(X_test_scaled)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 10))
sns.heatmap(confusion_matrix(y_test, lr_pred), annot=True, cmap='coolwarm',fmt='g', ax=ax[1]).set_title('Logistic Regression')
sns.heatmap(confusion_matrix(y_test, knn_pred), annot=True, cmap='coolwarm',fmt='g', ax=ax[0]).set_title('KNN Classifier')
plt.show()
```

```python
print('Logistic Regression: \n', classification_report(y_test, lr_pred))
print('KNN Classifier: \n', classification_report(y_test, knn_pred))
```

```python
models = ['Logistic Regression', 'KNN Classifier']
accuracy = [accuracy_score(y_test, lr_pred)*100, accuracy_score(y_test, knn_pred)*100]
sns.barplot(x=models, y=accuracy).set_title('Model Comparison')
plt.xticks(rotation=90)
plt.ylabel('Accuracy')
print(accuracy)
```
